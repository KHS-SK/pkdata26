### 패키지 내 모듈을 미리 import

```python
st = '''from .graphic.render import render_test

VERSION = 4.6

def print_version_info():
    print(f"The version of this game is {VERSION}.")

'''
with open("/home/khs/data/game/__init__.py", "w") as f:
    
    f.write(st)
```
