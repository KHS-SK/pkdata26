````python
# 필터용 함수 기본코드, 양수 만들기(공기청정기용 필터)

test_air = [1, 23, 32, -5, 45, 0, -3, 123, 2, -13]

def pos(num):
    result = []
    for n in num:
        if n > 0:
            result.append(n)
    return result

def neg(num):
    result = []
    for n in num:
        if n < 0:
            result.append(n)
    return result

print(pos(test_air))
print(neg(test_air))
````

````python
# 기본코드 응용 (리스트 컴프리헨션)

test_air = [1, 23, 32, -5, 45, 0, -3, 123, 2, -13]
print([n for n in test_air if n > 0])
````

````python
# filter ** 중요
# filter(함수, 반복_가능한_데이터)
# 위 함수와 대비하여 filter 내장함수 적용, 공기청정기용 필터 만들기 // 필터는 리턴이 True !! 가 중요!

test_air = [1, 23, 32, -5, 45, 0, -3, 123, 2, -13]

def pos_fn(n):
    return n > 0
    
def neg_fn(n):
    return n < 0

print(list(filter(pos_fn, test_air)))
print(list(filter(neg_fn, test_air)))
````

````python
# filter 응용

test_air = [1, 23, 32, -5, 45, 0, -3, 123, 2, -13]

print(list(filter(lambda x: x > 0, test_air)))
print(list(filter(lambda x: x < 0, test_air)))
````

````python
# map용 함수 기본코드, 가공

sample = [1, 2, 3, 4, 11, 22, 33, 44]

def two_times(numberList):
    result = []

    for number in numberList:
        result.append(number*2)
    return result

result = two_times(sample)
print(result)
````

````python
# map용 함수 기본코드 응용 (리스트 컴프리헨션)

sample = [1, 2, 3, 4, 11, 22, 33, 44]

print([n*2 for n in sample])
````

````python
# map ** 중요 / map(함수, 반복_가능한_데이터)

sample = [1, 2, 3, 4, 11, 22, 33, 44]

def two_times(x):
    return x*2

print(list(map(two_times, sample)))
````

````python
# map 응용 (lambda)

sample = [1, 2, 3, 4, 11, 22, 33, 44]

print(list(map(lambda x : x*2, sample)))
````

````python
# 위 논리에서 45가 진짜 나오지 않는지 검증하는 프로그램을 만들어라. (검증횟수: 1,000,000회)

import random as rr
box = []
n = 0
cc = int(input("몇 회?") or 1000000)

while n < cc:
    num = ((rr.random()*45) + 1)
    num += 1
    box.append(num)

if cc == len(box):
    print(f'{cc}회 정상 진행 되었습니다.')

    for n in range(0, 45):
        print(f'{n}이 {box.count(n)}번 출현하였습니다.')
else:
    print(f'검증 결과 이상이 있습니다')
````

````python
# functools.reduce

def add(data):
    result = 0

    for i in data:
        result += i
    return result

data = [1, 2, 3, 4, 5]
result = add(data)

print(result)
````

````python
import os
import glob
import zipfile

# 1. 파일 생성

f = os.system("ls temp")
if (f == 512):
    os.system("mkdir temp")
i = 0
path = "./temp/"
while ord('g') >= (97 + i):
    with open(path + "test_" + chr(97 + i)*2 + ".txt","w") as f:
        f.write("테스트 중입니다.")
    i += 1

# 2. 파일 확인

flist = glob.glob(path + '*')

# 3. 파일 압축

with zipfile.ZipFile(path + 'all.zip','w') as z:
    for f in flist:
        z.write(f)
````

## threading

````python
import time

def long_task():
    for i in range(5):
        time.sleep(1)
        print("working:%s\n" % i)

print("Start")

for i in range(5):
    long_task()

print("End")
````

````python
# 스레드 사용

import time
import threading
start = time.time()

def long_task():
    data = 0
    for i in range(50*1000*1000):
        # time.sleep(1)
        # print("working:%s\n" % i)
        data += i

print("Start")

# threads = []

threads = [threading.Thread(target=long_task) for i in range(20)]

# for i in range(5):
#     t = threading.Thread(target=long_task)
#     threads.append(t)

for t in threads:
    t.start()

for t in threads:
    t.join()

print("End")
print("threading 연산시간: ", time.time() - start)
````

````python
# 멀티코어

import time
import multiprocessing
start = time.time()

def long_task():
    data = 0
    for i in range(1000*1000*1000):
        # time.sleep(1)
        # print("working:%s\n" % i)
        data += i

print("Start")
        
processes = [multiprocessing.Process(target = long_task) for i in range(20)]

for p in processes:
    p.start()
    
for p in processes:
    p.join()

print("End")
print("threading 연산시간: ", time.time() - start)
````

````python
# urllib

import urllib.request

def get_wikidocs(page):
    resource = 'https://wikidocs.net/{0}'.format(page)
    with urllib.requests.urlopen(resource) as s:
        with open('wikidocs_%s.html' % page, 'wb') as f:
            f.write(s.read())
````

````python
from fractions import Fraction
import sympy

x = sympy.symbols("x")

f = sympy.Eq(x*Fraction('2/5'), 1760)

result = sympy.solve(f)

remains = result[0] - 1760

print('남은 돈은 {}원 입니다.'.format(remains))
````

````python
# 응용

import requests as req
from bs4 import BeautifulSoup as bs

url = "https://www.melon.com/chart/index.htm"
headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/145.0.0.0 Safari/537.36 Edg/145.0.0.0"}
web = req.get(url, headers=headers)
soup = bs(web.content, 'html.parser')

title = soup.select('.rank01')[:20]
name = soup.select('.checkEllipsis a')[:20]

str = ''
for i, (t, n) in enumerate(zip(title, name),1):
    str += f'{i}위: {t.text.strip()} / {n.text}\n'
print(str, end = '')
````
