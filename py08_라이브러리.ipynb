````python
# 필터용 함수 기본코드, 양수 만들기(공기청정기용 필터)

test_air = [1, 23, 32, -5, 45, 0, -3, 123, 2, -13]

def pos(num):
    result = []
    for n in num:
        if n > 0:
            result.append(n)
    return result

def neg(num):
    result = []
    for n in num:
        if n < 0:
            result.append(n)
    return result

print(pos(test_air))
print(neg(test_air))
````

````python
# 기본코드 응용 (리스트 컴프리헨션)

test_air = [1, 23, 32, -5, 45, 0, -3, 123, 2, -13]
print([n for n in test_air if n > 0])
````

````python
# filter ** 중요
# filter(함수, 반복_가능한_데이터)
# 위 함수와 대비하여 filter 내장함수 적용, 공기청정기용 필터 만들기 // 필터는 리턴이 True !! 가 중요!

test_air = [1, 23, 32, -5, 45, 0, -3, 123, 2, -13]

def pos_fn(n):
    return n > 0
    
def neg_fn(n):
    return n < 0

print(list(filter(pos_fn, test_air)))
print(list(filter(neg_fn, test_air)))
````

````python
# filter 응용

test_air = [1, 23, 32, -5, 45, 0, -3, 123, 2, -13]

print(list(filter(lambda x: x > 0, test_air)))
print(list(filter(lambda x: x < 0, test_air)))
````

````python
# map용 함수 기본코드, 가공

sample = [1, 2, 3, 4, 11, 22, 33, 44]

def two_times(numberList):
    result = []

    for number in numberList:
        result.append(number*2)
    return result

result = two_times(sample)
print(result)
````

````python
# map용 함수 기본코드 응용 (리스트 컴프리헨션)

sample = [1, 2, 3, 4, 11, 22, 33, 44]

print([n*2 for n in sample])
````

````python
# map ** 중요 / map(함수, 반복_가능한_데이터)

sample = [1, 2, 3, 4, 11, 22, 33, 44]

def two_times(x):
    return x*2

print(list(map(two_times, sample)))
````

````python
# map 응용 (lambda)

sample = [1, 2, 3, 4, 11, 22, 33, 44]

print(list(map(lambda x : x*2, sample)))
````

````python
# 위 논리에서 45가 진짜 나오지 않는지 검증하는 프로그램을 만들어라. (검증횟수: 1,000,000회)

import random as rr
box = []
n = 0
cc = int(input("몇 회?") or 1000000)

while n < cc:
    num = ((rr.random()*45) + 1)
    num += 1
    box.append(num)

if cc == len(box):
    print(f'{cc}회 정상 진행 되었습니다.')

    for n in range(0, 45):
        print(f'{n}이 {box.count(n)}번 출현하였습니다.')
else:
    print(f'검증 결과 이상이 있습니다')
````

````python
# itertools.zip_longest

import itertools

students = ["한민서", "황지민", "이영철", "이광수", "김승민"]
snacks = ["사탕", "초콜릿", "젤리"]

result = itertools.zip_longest(students, snacks, fillvalue = "새우깡")

print(list(result))
````

````python
# functools.reduce

def add(data):
    result = 0

    for i in data:
        result += i
    return result

data = [1, 2, 3, 4, 5]
result = add(data)

print(result)
````
