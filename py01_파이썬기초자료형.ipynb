## 사칙연산

```python
# 사칙연산 예

sa = 77
san = 13
print('덧셈:',sa + san)
print('뺄셈:',sa - san)
print('곱셈:',sa * san)
print('나눗셈:',sa / san)
print('몫:',sa // san) # 좀 중요
print('나머지:',sa % san) # 좀 중요
print('제곱:',sa ** san)
```

```python
# 문제 / input 으로 개수를 입력받아 

# 한잔에 2500원인 아이스 아메리카노와 3800원인 아이스 카페라떼를 계산한 영수증을 출력하라

# 예
'''
====================
       영수증
--------------------
아이스아메리카노 7잔 (가운데 정렬)
아이스카페라떼 4잔
--------------------


>> 총가격:      원 (우측정렬)
====================
'''
```

```python
A = 2500 # 아이스아메리카노
B = 3800 # 아이스카페라떼

num1 = input("몇잔의 아이스아메리카노를 마시겠습니까?")
num2 = input("몇잔의 아이스카페라떼를 마시겠습니까?")

rec = "영수증"

print("=" * 20)
print("{0:^17}".format("영수증"))
print("-" * 20)
print("아이스아메리카노 %d잔" % int(num1))
print("아이스카페라떼 %d잔" % int(num2))
print("-" * 20)
print("\n\n {0:>6}: {1:>6}원".format("총가격", int(A * int(num1) + B * int(num2))))
print("=" * 20)
```
