### 교집합, 합집합, 차집합 구하기

````python
s1 = set([1, 2, 3, 4, 5, 6])
s2 = set([4, 5, 6, 7, 8, 9])

# 교집합 / 기호: &

print("교집합", s1 & s2) # s1.intersection(s2) 가능

# 합집합 / 기호: |
print("합집합", s1 | s2) # s1.union(s2) 가능

# 차집합 / 기호: -
print("차집합", s1 - s2, s2 - s1) # s1.difference(s2), s2.difference(s1) 가능
````

### 집합 자료형 관련 함수

````python
# add 집합 자료형 값 1개 추가

s1 = set ([1, 2, 3])
print(s1)

s1.add(4)
print(s1)
````

````python
# update 집합 자료형 값 여러개 추가

s1 = set ([1, 2, 3])
print(s1)

s1.update([2, 3, 4, 5, 6, 7])
print(s1)
````

````python
# remove 집합 자료형 특정 값 제거하기 

s1 = set ([1, 2, 3])
print(s1)

s1.remove(2) # index가 아닌 실제 문자열 값을 선택
print(s1)
````

**존재하지 않는 값을 찾을 땐 오류**



````python
# discard 집합 자료형 특정 값 제거하기 (존재하지 않는 값을 제거하려 해도 오류가 발생하지 않음)

s1 = set ([1, 2, 3])
print(s1)

s1.discard(0) # 없는 값이어도 오류가 발생하지 않음
print(s1)
````

````python
# clear 집합 자료형 내 모든 값 제거

s1 = set([1, 2, 3])
s1.clear()
print(s1)
````

### 변수를 만드는 여러가지 방법

````python
(a, b) = (123, 456)

print(a)
print(b)
````

````python
[a, b, c] = ['python', 'life', 'cool']

print(a)
print(b)
print(c)
````

````python
a = b = c = 0

print(a)
print(b)
print(c)
````

````python
# 두 변수의 값을 바꿀 때

a = 33
b = 55
print(a, b)

a, b = b, a
print(a, b)
````

````python
# 다른 언어 예 temp

aa = 33
bb = 55
print(aa, bb)

temp = aa
aa = bb
bb = temp

print(aa, bb)
````
