```python
def counter_maker():
    count = [0]

    def counter():
        count[0] += 1
        return count[0]

    return counter

c1 = counter_maker()
c2 = counter_maker()
```

```python
import time

def myfunc():
    start = time.time()
    print("함수가 실행됩니다.")
    end = time.time()
    print("함수 수행시간: %f 초" % (end-start))

myfunc()
```

```python
import time

def outfunc(X):
    def infunc():
        start = time.time()
        print("함수가 준비.")
        result = X()
        print("함수 수행시간: %f 초" % (time.time()-start))
        return result
    return infunc

@outfunc
def Xfunc():
    print("함수가 실행됩니다.")
@outfunc
def Yfunc():
    print("함수가 어렵습니다.")
```

```python
# a = outfunc(Xfunc)
# b = outfunc(Xfunc)
# c = outfunc(Xfunc)
# a()
# b()
# c()

Xfunc()
Yfunc()
```

```python
Xfunc = outfunc(Xfunc)
Yfunc = outfunc(Xfunc)

Xfunc()
Yfunc()
```

```python
import time

def elapsed(original_func):
    def wrapper():
        start = time.time()
        result = original_func()
        end = time.time()
        print("함수 수행 시간: %f초" % (end - start))
        return result
    return wrapper

def myfuc():
    print("함수가 실행됩니다.")

decorated_myfunc = elapsed(myfunc)
decorated_myfunc()
```

```python
# 이터레이터 만들기

class MyIterator:
    def __init__(self, data):
        self.data = data
        self.position = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.position >= len(self.data):
            raise StopIteration
        result = self.data[self.position]
        self.position += 1
        return result

if __name__ == "__main__":
    i = MyIterator([1, 2, 3])
    for item in i:
        print(item)
```

```python
it = MyIterator([1, 2, 3])
```

```python
for n in it:
    print(n)
```

```python
# 제너레이터

def mybot():
    print("작업시작")
    yield '1번 작업 완료'
    print("로봇 A 위치로 동작")
    yield '2번 작업 완료'
    print("로봇 A 그리퍼 동작 후 위치 반환")
    yield '3번 작업 완료'
    print("로봇 B 가동 시작")
    yield "end"

gb = mybot()
```

```python
try:
    next(gb)
except StopIteration as e:
    print("더 이상 할 작업이 없습니다.", e)
```
