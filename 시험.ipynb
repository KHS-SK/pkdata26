```python
# Q4  - 온도 데이터를 가져오는 코드이다. 빠진 부분을 넣으시오.

# 현재 온도 출력
import requests as req

url = "https://api.open-meteo.com/v1/forecast?latitude=35.116741&longitude=129.089540&current_weather=true"
value = req.get(url).json()

tem = value['current_weather']['temperature']
unit = value['current_weather_units']['temperature']

print(f"현재 온도: {tem}{unit}")
```

```python
# Q5 - 다음 문제의 출력이 나오도록 리스트 컴프리핸션을 작성하라.

name = ["홍길동", "이순신", "슈퍼맨", "배트맨", "박서방"]
age = [22, 33 ,34, 12, 54]

# 딕셔너리로 만들기

print(dict([(y, x) for y, x in zip(name, age)]))
```

```python
# Q6 - 다음은 배열속의 튜플을 각각 name과 age, grade 변수에 담는 내용이다. 빈칸에 list comprehension 으로 해결되는 코드를 넣어라.

students = [("jane", 22, "A"), ("dave", 32, "B"), ("sally", 17, "B")]

[name, age, grade] = [(x, y, z) for x, y, z in students]

print(name, age, grade, sep = '\n')
```

```python
# Q7 - 다음은 제너레이터로 파일 속의 에러 내용을 반환하는 로직이다. 빈칸에 내용을 넣으시오.

def error_lines(filepath):
    with open(filepath, "r", encoding = "utf-8") as f:
        for line in f:
            if "ERROR" in line:
                 line.strip()

# 사용 예시

for err in error_lines("app.log"):
    print("에러 발견:", err)
```
