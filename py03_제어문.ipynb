````python
# 응용(6개의 눈금 주사위 n번 던져서 눈금 출현 확률 구하기)

import random as ran

n = 0
box =[]
cc = int(input("몇회 주사위를 던질까요?: ") or 100)

while n < cc:
    num = ran.randint(1, 6)
    n += 1
    box.append(num)
    # print(f'{n}번쨰:', num)

if cc == len(box):
    print(f"{cc}회 정상 진행 되었습니다.")
    print(f"주사위 눈금 1이 {box.count(1)}번 등장하였습니다. 주사위 눈금 1이 나온 확률은 {box.count(1)/cc*100:.2f}%입니다.")
    print(f"주사위 눈금 2이 {box.count(2)}번 등장하였습니다. 주사위 눈금 2이 나온 확률은 {box.count(2)/cc*100:.2f}%입니다.")
    print(f"주사위 눈금 3이 {box.count(3)}번 등장하였습니다. 주사위 눈금 3이 나온 확률은 {box.count(3)/cc*100:.2f}%입니다.")
    print(f"주사위 눈금 4이 {box.count(4)}번 등장하였습니다. 주사위 눈금 4이 나온 확률은 {box.count(4)/cc*100:.2f}%입니다.")
    print(f"주사위 눈금 5이 {box.count(5)}번 등장하였습니다. 주사위 눈금 5이 나온 확률은 {box.count(5)/cc*100:.2f}%입니다.")
    print(f"주사위 눈금 6이 {box.count(6)}번 등장하였습니다. 주사위 눈금 6이 나온 확률은 {box.count(6)/cc*100:.2f}%입니다.")
else:
    print(f"검증 결과 이상이 있습니다. {cc}회 목표로 동작하였으나, {len(box)}으로 측정되었습니다.")
````

````python
# random으로 스토리 생성기 만들기

import random as ran

ran.seed(7)

characters = ["용감한 기사", "모험심 넘치는 고양이", "호기심 많은 로봇", "신나게 출근한 신영이", "서투른 마법사", "명석한 탐정", "즐거운 선생님"]
places = ["어두운 숲에서", "먼 우주 행성에서", "마법의 성에서", "엄중한 교실에서", "바닷속 도시에서", "귀신이 나오는 흉가에서", "썸남과 엘베안에서"]
events = ["숨겨진 보물을 발견했다", "무시무시한 용과 싸웠다", "비밀 통로를 발견했다", "고대의 미스터리를 풀었다", "왕국을 위험에서 구하였다", "결혼날짜를 잡았다"]

ch = ran.choice(characters)
pl = ran.choice(places)
ev = ran.choice(events)
ch2 = ran.choice(characters)
pl2 = ran.choice(places)
ev2 = ran.choice(events)

print("어느 날, %s는 %s 모험을 떠나 %s 그리고 %s는 %s %s" %(ch, pl, ev, ch2, pl2, ev2))
````

````python
# for와 range를 이용한 구구단

# 2단

x = 2

print('=' * 14)
print("{0:^14}".format("2단"))
print('-' * 14)


for i in range(1, 10):

    print(x, "x", i, "=", x * i)

print('=' * 14)
````

````python
# for와 range를 이용한 구구단

# 구구단 전체
print("=" * 10)
for i in range(2, 10):
    print("{0:>5}{1:<5}".format(i,"단"),)
    print('-' * 10)
    for j in range(1,10):
        print(i, "x", j, "=", i * j)
    print('-' * 10)
print("=" * 10)
````

````python
# 문제풀이

for j in range(2, 10):
    print(f"{(str(j) + "단"):^12}")
    print("-" * 12)
    for i in range(1, 10):
        print(f"{j} x {i} = {(j * i):>2}")
    print("=" * 12)
````

### zip 함수로 여러 리스트 함께 순회하기

````python
names = ['홍길동', '김철수', '이영희']
korean = [85, 92, 78]
english = [90, 88, 95]
for name, kor, eng in zip(names, korean, english):
    print(f'{name}: 국어 {kor} 점, 영어 {eng} 점')
````

````python
names = ['홍길동', '김철수', '이영희', '이순신']
scores = [85, 92, 78, 99]

# grade = ["합격" for x in scores if x > 85] // if 조건이 되면 합격이라는 결과를 보냄

grade = ["합격" if x > 85 else "불합격" for x in scores] 

for x, y, z in zip(names, scores, grade):
    print(f'{x}님은 {y}점 이므로 {z}입니다.')
````

````python
import random as ran
import matplotlib.pyplot as plt

box = []
n = 0
cc = 100000
values = []
choice = int(input("조작할 확률 %를 입력하세요") or 10) # 퍼센트 조절
my = int(input("조작할 번호를 입력하세요") or 3)

while n < cc:
    num = ran.randint(1, 6)
    lucky = ran.randint(1, 100)
    if choice >= lucky:
        box.append(my)
    else:
        box.append(num)
    n += 1

if cc == len(box):
    print(f'{cc}회 정상 진행 되었습니다.')

    for n in range(1, 7):
        print(f'주사위 눈금 {n}이 {box.count(n)}번 { ((box.count(n)/cc) * 100): 0.2f}% 출현하였습니다.')
        values.append(box.count(n))
else:
    print(f'검증 결과 이상이 있습니다. {cc}회 목표로 동작하였으나, {len(box)}으로 측정 되었습니다.')


color = ["skyblue", "orange", "skyblue", "orange", "skyblue", "orange"]
plt.bar(range(1,7), values, color=color)
plt.show()
````
