## threading

```python
import time

def long_task():
    for i in range(5):
        time.sleep(1)
        print("working:%s\n" % i)

print("Start")

for i in range(5):
    long_task()

print("End")
```

```python
# 스레드 사용

import time
import threading
start = time.time()

def long_task():
    data = 0
    for i in range(50*1000*1000):
        # time.sleep(1)
        # print("working:%s\n" % i)
        data += i

print("Start")

# threads = []

threads = [threading.Thread(target=long_task) for i in range(20)]

# for i in range(5):
#     t = threading.Thread(target=long_task)
#     threads.append(t)

for t in threads:
    t.start()

for t in threads:
    t.join()

print("End")
print("threading 연산시간: ", time.time() - start)
```

```python
# 멀티코어

import time
import multiprocessing
start = time.time()

def long_task():
    data = 0
    for i in range(1000*1000*1000):
        # time.sleep(1)
        # print("working:%s\n" % i)
        data += i

print("Start")
        
processes = [multiprocessing.Process(target = long_task) for i in range(20)]

for p in processes:
    p.start()
    
for p in processes:
    p.join()

print("End")
print("threading 연산시간: ", time.time() - start)
```

```python
# 3의 배수의 합 구하기

result = 0
i = 1

while i <= 1000:
    if i % 3 == 0:
        result += i
    i += 1

print(result)
```

```python
# 평균 점수 구하기

score = [70, 60, 55, 75, 95, 90, 80, 80, 85, 100]
total = 0

for A in score:
    total += A
average = total / len(score)

print(average)
```

```python
# 리스트 컴프리헨션 사용하기


numbers = [1, 2, 3, 4, 5]
result = []

for n in numbers:
    if n % 2 == 1:
        result.append(n * 2)

print(result)
```

```python
numbers = [1, 2, 3, 4, 5]
result = [num * 2 for num in numbers if num % 2 == 1]
print(result)
```
