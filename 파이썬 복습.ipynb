## threading

```python
import time

def long_task():
    for i in range(5):
        time.sleep(1)
        print("working:%s\n" % i)

print("Start")

for i in range(5):
    long_task()

print("End")
```

```python
# 스레드 사용

import time
import threading
start = time.time()

def long_task():
    data = 0
    for i in range(50*1000*1000):
        # time.sleep(1)
        # print("working:%s\n" % i)
        data += i

print("Start")

# threads = []

threads = [threading.Thread(target=long_task) for i in range(20)]

# for i in range(5):
#     t = threading.Thread(target=long_task)
#     threads.append(t)

for t in threads:
    t.start()

for t in threads:
    t.join()

print("End")
print("threading 연산시간: ", time.time() - start)
```

```python
# 멀티코어

import time
import multiprocessing
start = time.time()

def long_task():
    data = 0
    for i in range(1000*1000*1000):
        # time.sleep(1)
        # print("working:%s\n" % i)
        data += i

print("Start")
        
processes = [multiprocessing.Process(target = long_task) for i in range(20)]

for p in processes:
    p.start()
    
for p in processes:
    p.join()

print("End")
print("threading 연산시간: ", time.time() - start)
```
