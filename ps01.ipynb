````python
# 비트코인 API

import requests as req

url = "https://api.bithumb.com/v1/candles/minutes/1?market=KRW-BTC&count=1"
d_url = "https://api.bithumb.com/v1/candles/days?market=KRW-BTC&?count=1"

ori = req.get(url).json()[0]
ori2 = req.get(d_url).json()[0]

# * 분봉
# 시작값
open_p = float(ori['opening_price'])
# 최고값
max_p = float(ori['high_price'])
# 최저값 
min_p = float(ori['low_price'])
# 변화폭=\
==-=
band_p = max_p - min_p

# * 일봉
# 시작값
open_p2 = float(ori['opening_price'])
# 최고값
max_p2 = float(ori['high_price'])
# 최저값 
min_p2 = float(ori['low_price'])
# 변화폭
band_p2 = max_p2 - min_p2

print("분기준 시작값:", open_p, "\n최고값:", max_p, "\n최저가:", min_p, "\n변화폭:", band_p)
print("일기준 시작값:", open_p2, "\n최고값:", max_p2, "\n최저가:", min_p2, "\n변화폭:", band_p2)

# print("검토: ", open_p + band_p)
# print("최고값: ", max_p)

# 파이썬 외 언어, 삼항 연산자 / print("최종: ", ((max_p - (open_p + band_p)) > 0) ?  "떡상 중" : "떡락 중"  )

result = "떡상중" if ((max_p - (open_p + band_p)) > 0) else "떡락 중"
result = "떡상중" if ((max_p2 - (open_p2 + band_p2)) > 0) else "떡락 중"

print("분봉 판단:", result)
print("일봉 판단:", result)
````
